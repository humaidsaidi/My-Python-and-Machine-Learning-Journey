di sini kita memiliki dataset yang memiliki features company, job, pendidikan, dan apakah gajinya lebih dari 100k dolar.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/salaries.csv')
df
```

Selanjutnya kita akan membuat decision tree untuk menentukan karyawan seperti apakah yang mendapat gaji di atas 1000 dolar.

```python
# mesisahkan antara variabel independen(input) dan variabel target
inputs = df.drop('salary_more_then_100k', axis=1)
target = df['salary_more_then_100k']
```

```python
inputs
```

```python
target
```

karena algoritma machine learning hanya faham data yang berbentuk angka, maka kita harus merubah data kita yang berbentuk string ke bentuk angka. Di sini kita menggunakan metode label encoder untuk melakukan hal tersebut.

```python
from sklearn.preprocessing import LabelEncoder

# karana ada 3 kolom yang ingin dirubah ke bentuk angka maka kita buat 3 objek dari kelas LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
# menambah kolom baru pada dataframe yang isinya adalah hasil encoding setiap variabel independent
inputs['company_n'] = le_company.fit_transform(inputs['company']) 
inputs['job_n'] = le_company.fit_transform(inputs['job']) 
inputs['degree_n'] = le_company.fit_transform(inputs['degree'])

inputs
```

```python
# membuat dataframe baru yang isinya hanya hasil encoding dari dataframe inputs
inputs_n = inputs.drop(['company', 'job', 'degree'], axis=1)
inputs_n
```

company = <br>
- jika nilainya 0 artinya adalah abc pharma
- jika nilainya 1 artinya adalah facebook
- jika nilainya 2 artinya adalah google

job = <br>
- jika nilainya 0 artinya adalah bussines manager
- jika nilainya 1 artinya adalah computer programmer
- jika nilainya 2 artinya adalah google sales executive

degree = <br>
- jika nilainya 0 artinya adalah bachelor
- jika nilainya 1 artinya adalah master

```python
# training classifier
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
model.fit(inputs_n, target)
```

```python
# mengecek skor dari model
model.score(inputs_n, target)
```

```python
# memprediksi gaji karyawan google di bidang sales executive sales dan berpendidikan master
model.predict([[2,2,1]])
```

karena outputnya 0 maka jawabannya gajinya di bawah 1000 dolar

```python
model.predict([[2,0,1]])
```

dengan menggunakan dataset yang diberikan tentukan apakah penumpang akan selamat atau tidak.

```python
df_t = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/Exercise/titanic.csv')
df_t
```

```python
# data cleaning (kolom age ada yang datanya kosong)
df_t['Age'] = df_t['Age'].fillna(df_t['Age'].mean())
```

```python
# mencari variabel independent mana yang paling mempengaruhi target
df_t.groupby('Survived').mean()
```

nampaknya variabel fare adalah yang paling menentukan variabel target

```python
inputs_t = df_t.drop(['PassengerId', 'Survived', 'Name', 'SibSp','Parch', 'Ticket', 'Cabin', 'Embarked'],axis=1)
target_t = df_t.Survived
```

```python
inputs_t 
```

```python
target_t
```

```python
from sklearn.preprocessing import LabelEncoder

# hanya satu kolom yang akan di label encoding
le_sex = LabelEncoder()
```

```python
# kolom sex isinya diganti dengan hasil encoding
inputs_t['Sex'] = le_sex.fit_transform(inputs_t['Sex']) 
```

```python
# training classifier
from sklearn import tree
```

```python
model_t = tree.DecisionTreeClassifier()
model_t.fit(inputs_t, target_t)
```

```python
model_t.score(inputs_t, target_t)
```
