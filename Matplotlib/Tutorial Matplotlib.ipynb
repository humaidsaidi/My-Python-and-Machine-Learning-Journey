```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Grafik Dasar

```python
x = [1,2,3]
y = [4,5,6]

plt.plot(x,y)
#plt.title('Grafik Pertama Saya') #memberi judul pada grafik
plt.title('Grafik Pertama Saya', fontdict={'fontname': 'Comic Sans MS', 'fontsize':20}) #memberi judul, font dan ukuran font
plt.xlabel('Sumbu X') #memberi label pada sumbu x
plt.ylabel('Sumbu Y') #memberi label pada sumbu y

plt.show() #bila tidak diberi ini nanti akan muncul '[<matplotlib.lines.Line2D at 0x7fee809cbf40>]' di atas grafik
```

```python
#agar angka pada sumbu menjadi integer dan tidak float

x = [1,2,3]
y = [4,5,6]

plt.plot(x,y)
#plt.title('Grafik Pertama Saya') #memberi judul pada grafik
plt.title('Grafik Pertama Saya', fontdict={'fontname': 'Comic Sans MS', 'fontsize':20}) #memberi judul, font dan ukuran font
plt.xlabel('Sumbu X') #memberi label pada sumbu x
plt.ylabel('Sumbu Y') #memberi label pada sumbu y

plt.xticks(x) #agar angka pada sumbu y menjadi integer dan tidak menjadi float
plt.yticks(y)

plt.show() #bila tidak diberi ini nanti akan muncul '[<matplotlib.lines.Line2D at 0x7fee809cbf40>]' di atas grafik
```

```python
#membuat label pada garis grafik, sekaligus menentukan warna grafik, ukuran tebal, dan tanda setiap persinggungan
#mirip legenda pada peta

x = [1,2,3]
y = [4,5,6]

#memberi label dan mengganti warna garis grafik, menentukan tebal garis, dan memberi tanda pada persinggungan
plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', markersize=10) 
#plt.plot(x,y, 'r,--', label='2x') #versi pendek dari kode di atas

#plt.title('Grafik Pertama Saya') #memberi judul pada grafik
plt.title('Grafik Pertama Saya', fontdict={'fontname': 'Comic Sans MS', 'fontsize':20}) #memberi judul, font dan ukuran font
plt.xlabel('Sumbu X') #memberi label pada sumbu x
plt.ylabel('Sumbu Y') #memberi label pada sumbu y

plt.legend() #untuk mengaktifkan legenda

plt.show() #bila tidak diberi ini nanti akan muncul '[<matplotlib.lines.Line2D at 0x7fee809cbf40>]' di atas grafik
```

```python
#mengganti warna titik penanda persingguhan garis grafik, sekaligus membuat linestyle

x = [1,2,3]
y = [4,5,6]

#memberi label dan mengganti warna garis grafik, menentukan tebal garis, dan memberi tanda pada persinggungan
plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue') 

#plt.title('Grafik Pertama Saya') #memberi judul pada grafik
plt.title('Grafik Pertama Saya', fontdict={'fontname': 'Comic Sans MS', 'fontsize':20}) #memberi judul, font dan ukuran font
plt.xlabel('Sumbu X') #memberi label pada sumbu x
plt.ylabel('Sumbu Y') #memberi label pada sumbu y

plt.legend() #untuk mengaktifkan legenda

plt.show() #bila tidak diberi ini nanti akan muncul '[<matplotlib.lines.Line2D at 0x7fee809cbf40>]' di atas grafik
```

```python
#membuat grafik yang memiliki 2 garis (ada 2 representasi data)
x = [1,2,3]
y = [4,5,6]

#garis kedua
x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2, x2**2, label='x^2', color='blue', linewidth=2, marker='.', markersize=10)

plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', markersize=10) 


plt.title('Grafik Pertama Saya', fontdict={'fontname': 'Comic Sans MS', 'fontsize':20}) 
plt.xlabel('Sumbu X') #memberi label pada sumbu x
plt.ylabel('Sumbu Y') #memberi label pada sumbu y

plt.legend() 

plt.show() 
```

```python
#misalnya kita ingin membatasi garis hanya sampai rentang tertentu, dan selebihnya garis putus2
#membuat grafik yang memiliki 2 garis (ada 2 representasi data)
x = [1,2,3]
y = [4,5,6]

#garis kedua
x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2[:6], x2[:6]**2, label='x^2', color='blue', linewidth=2, marker='.', markersize=10) #hanya 5 awal dibutuhkan
plt.plot(x2[5:], x2[5:]**2, 'b--')#selebihnya garis putus2

plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', markersize=10) 


plt.title('Grafik Pertama Saya', fontdict={'fontname': 'Comic Sans MS', 'fontsize':20}) 
plt.xlabel('Sumbu X') 
plt.ylabel('Sumbu Y') 

plt.legend() 

plt.show() 
```

### Grafik Batang (Bar Chart)

```python
label = ['A', 'B', 'C']
nilai = [1,4,2]

plt.bar(label, nilai, color='y') #membuat bar chart

plt.figure(figsize=(6,4))

plt.show()
```

```python
#memberi pola di dalam bar

label = ['A', 'B', 'C']
nilai = [1,4,2]

bar = plt.bar(label, nilai, color='y') #membuat bar chart

#bar[0].set_hatch('/') #memberi pola di dalam bar
#bar[1].set_hatch('O')
#bar[2].set_hatch('*')

#versi menggunakan perulangan for
pola = ['/','O','*']
for b in bar:
    b.set_hatch(pola.pop(0))

plt.figure(figsize=(6,4))

plt.show()
```

```python
#menyeling setiap 3 tahun

gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))
plt.title('Harga Gas Sepanjang Waktu dalam Dolar', fontdict={'fontname': 'Comic Sans MS', 'fontsize':20})

plt.plot(gas.Year, gas.USA, label='USA', color='red', linewidth=2, marker='.')
plt.plot(gas.Year, gas.UK, label='UK', color='blue', linewidth=2, marker='.')
plt.plot(gas.Year, gas.Australia, label='Australia', color='orange', linewidth=2, marker='.')
plt.plot(gas.Year, gas['South Korea'], label='South Korea', color='green', linewidth=2, marker='.')

plt.xlabel('Tahun', fontdict={'fontname': 'Comic Sans MS', 'fontsize':15})
plt.ylabel('Dolar', fontdict={'fontname': 'Comic Sans MS', 'fontsize':15})

plt.xticks(gas.Year[::3])

plt.legend()
plt.show()
```

```python
#grafik seluruh negara

gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(16,10))
plt.title('Harga Gas Sepanjang Waktu dalam Dolar', fontdict={'fontname': 'Comic Sans MS', 'fontsize':20})

colors=['red','brown','yellow', 'blue', 'magenta', 'green', 'purple', 'pink', 'cyan', 'olive' ]

for country, color in zip(gas, colors):
    if country != 'Year':
        plt.plot(gas.Year, gas[country], label=country, color=color, linewidth=2, marker='.')

plt.xlabel('Tahun', fontdict={'fontname': 'Comic Sans MS', 'fontsize':15})
plt.ylabel('Dolar', fontdict={'fontname': 'Comic Sans MS', 'fontsize':15})

plt.xticks(gas.Year)

plt.legend()
plt.show()
```

### Contoh Lain (FIFA DATA)

```python
fifa = pd.read_csv('fifa_data.csv')
fifa
```

#### Membuat Histogram

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins, color='#5df542') #untuk menentukkan warna, gunakan color picker dan masukkan kode warnanya
plt.xticks(bins)
plt.xlabel('Skill Level')
plt.ylabel('Number of Players')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.show()
```

### Diagram Lingkaran (Pie Chart)

```python
fifa['Preferred Foot']
```

```python
#diagram yang menunjukkan banyak pemain yang menggunakan kaki kanan atau kiri

left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0] #menghitung banyak pemain yang menggunakan kaki kanan
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels= ['Left', 'Right']
colors = ['#f542e3', '#4245f5']

plt.pie([left, right], labels=labels, colors=colors, autopct='%.2f')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
lightMedium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
mediumHeavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count()[0]

plt.figure(figsize=(16,10))

weight = [light, lightMedium, medium, mediumHeavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']

plt.title('Weigt Distribution of FIFA Players (in lbs)')
plt.pie(weight, labels=labels, autopct='%.2f %%', pctdistance=0.8)
plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
lightMedium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
mediumHeavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count()[0]

plt.figure(figsize=(16,10))

weight = [light, lightMedium, medium, mediumHeavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (.1,.1,.1,.1,.1,)

plt.title('Weigt Distribution of FIFA Players (in lbs)')
plt.pie(weight, labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)
plt.show()
```

### Box Chart

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
juve = fifa.loc[fifa.Club == 'Juventus']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'Juventus']
plt.boxplot([barcelona, madrid, juve], labels=labels)

plt.ylabel('FIFA Overall Rating')
plt.title('Profesional Soccer Team Comparison')
plt.show()
```

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
juve = fifa.loc[fifa.Club == 'Juventus']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'Juventus']
boxes = plt.boxplot([barcelona, madrid, juve], labels=labels, patch_artist=True, medianprops={'linewidth':2})

for box in boxes['boxes']:
    # mengatur warna tepi box dan tebal garisnya
    box.set(color='#4245f5', linewidth=2)
    
    #Mengisi warna
    box.set(facecolor='#42f5e0')

plt.ylabel('FIFA Overall Rating')
plt.title('Profesional Soccer Team Comparison')
plt.show()
```
