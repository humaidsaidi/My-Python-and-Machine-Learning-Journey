```python
#Membuat fungsi untuk menyelesaikan permasalahan di atas dengan algoritma linear search
def locate_card(cards, query):
    #Membuat variabel dengan position dengan nilai 0
    position = 0
    
    #Membuat loop untuk pengulangan
    while position < len(cards):
        
        #Mengecek apakah pada posisi ini index elemen cocok dengan angka yang ditanya
        if cards[position] == query:
            
            #Jawaban ditemumkan!! Kembalikan nilai dan keluar dari fungsi
            return position
        
        #Increment variabel position
        position += 1
        
        #Mengecek apakah kita sudah mencapai penghujung dari list
        if position == len(cards):
            
            #Angka yang ditanya tidak ditemukan, Kembalikan nilai -1
            return -1
        
        
```

Pencarian kartu di atas masih menggunakan linear search yang mengharuskan kita untuk mengambil kartu satu persatu
dari awal, sehingga kemungkinan tebakan benar dengan sesedikit mungkin membuka kartu menjadi lebih kecil. Maka untuk
mengatasinya kita menggunakan algoritma binary search.

```python
#Fungsi untuk mengatasi permasalahan di atas dengan algoritma binary search
def locate_card(cards, query):
    lo, hi = 0, len(cards)-1
    
    while lo <= hi:
        mid = (lo + hi) // 2
        mid_number = cards[mid]
        
        print('lo:',lo, ' hi:',hi, ' mid:',mid, ' mid number:', mid_number)
        
        if mid_number == query:
            return mid
        elif mid_number < query:
            hi = mid - 1
        elif mid_number > query:
            lo = mid + 1
            
    return -1
```
